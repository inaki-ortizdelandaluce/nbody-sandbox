### 3.2 Forward Euler Integrator

$$
\textbf{r}(t+\Delta t)=\textbf{r}(t)+\dot{\textbf{r}}(t)\Delta t\\ 
\dot{\textbf{r}}(t+\Delta t)=\dot{\textbf{r}}(t)+\ddot{\textbf{r}}(t)\Delta t\\
\\
\ddot{\textbf{r}}(t)=-\frac{GM}{r^3(t)}\textbf{r}(t)\\
\mbox{ Force equals centrifugal acceleration:} \frac{GM}{r^2}=\frac{\dot{r}^2}{r} \Rightarrow \dot{r}=\sqrt{\frac{GM}{r}}\\
$$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

#### Constants

```python
M = 1
m = 3e-06
G = 1
```

#### Initial conditions

```python
X = np.array([0, 0, 0])  # Sun initial position
V = np.array([0, 0, 0])  # Sun initial velocity
x = np.array([1., 0, 0])  # Earth initial position
v = np.array([0, 1., 0])  # Earth initial velocity
```

```python
dt = 0.1
t_end = 100
x_vec = []
y_vec = []
X_vec = []
Y_vec = []

x_vec.append(x[0])
X_vec.append(X[0])
y_vec.append(x[1])
Y_vec.append(X[1])


for t in np.linspace(0,t_end,int(t_end/dt)):
    a = - G * M * (x - X) / np.linalg.norm(x - X)**3
    A = - G * m * (X - x) / np.linalg.norm(X - x)**3
    
    x = x + v * dt
    X = X + V * dt
    v = v + a * dt
    V = V + A * dt
    
    x_vec.append(x[0])
    X_vec.append(X[0])
    y_vec.append(x[1])
    Y_vec.append(X[1])

plt.plot(x_vec, y_vec)
plt.plot(X_vec, Y_vec)
plt.title('$M=m=1, \Delta t=0.0001$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.axis('equal')
plt.show()
```
